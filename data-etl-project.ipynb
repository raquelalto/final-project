```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#store csv into dataframe 
game_skater_df = pd.read_csv(r'C:\Users\quila\Desktop\ETL-Project\Resources\game_skater_stats.csv')
game_skater_df.head(10)
```

```python
#create new dataframe with select columms
new_game_skater_df = game_skater_df[['player_id', 'timeOnIce', 'assists', 'goals', 'shots',]].copy()
new_game_skater_df .head()
```

```python
#store second csv into dataframe 
player_info_df = pd.read_csv(r'C:\Users\quila\Desktop\ETL-Project\Resources\player_info.csv')
player_info_df.head(10)
```

```python
#create new dataframe with select columms
new_player_info_df = player_info_df [['player_id', 'firstName', 'lastName', 'nationality']].copy()
new_player_info_df .head()
```

```python
# username:password@localhost:5432/tablename
rds_connection_string = "admin1:12345@localhost:5432/NHL_Data"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
#load csv converted DataFrames into database
new_game_skater_df.to_sql(name='player_stats1', con=engine, if_exists='append', index=False)
new_player_info_df.to_sql(name='player_info1', con=engine, if_exists='append', index=False)
```

```python
#Confirm data has been added by querying tables
pd.read_sql_query('select * from player_stats1', con=engine).head()
```
